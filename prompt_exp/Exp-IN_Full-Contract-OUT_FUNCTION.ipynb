```python
import os
import openai
import backoff

# Setup OpenAI config

openai.api_key = os.getenv("OPENAI_API_KEY")
def sendAICall(system_prompt, user_prompt, temp=0, model="gpt-3.5-turbo-0125"):
    response = openai.ChatCompletion.create(
      model=model,
      messages=[
           {"role": "system", "content": system_prompt},
          {"role": "user", "content": user_prompt}
      ],
      temperature=temp,
      #max_tokens=1024
    )
    return response
```
